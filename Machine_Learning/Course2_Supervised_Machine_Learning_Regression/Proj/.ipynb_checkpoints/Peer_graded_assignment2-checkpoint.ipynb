```python
# 2. Data Cleaning 
# import os

import pandas as pd

path = 'data/GS.csv'
df = pd.read_csv(path)
df.head()
df.describe()
df.info() 
# no null data, no need to data cleaning
```

```python
# 3. Exploratory Data Analysis
# Check overall stock price visually

import matplotlib.pyplot as plt
import numpy as np

filter = ['Close']
y = df.loc[:, filter]

plt.plot(y)
plt.show()
```

```python
# 4. Feature Engineering and Variable Transformation
# We need only Open, High, Close, and Volume features.
# So we will delete unnecessary features.

needs = ['Open', 'High', 'Close', 'Volume']
y = df.loc[:,needs]
y
```

```python
# The scale of volume is too high, so we will standardize the scale of the volume feature.

import scipy.stats as st

y['Volume'] = st.zscore(df['Volume'])
y
```

```python
# Extract the features that we are interested in.

# 1. Find the Rise value of the day
x = y.loc[:, 'High'] - y.loc[:,'Open']
x = x.to_frame().rename(columns={0:'Rise'})

# 2. Find the change of Close price compared to yesterday
y.shape[0]
y1 = y.copy()
for i in range(1, y.shape[0]):
    y1.loc[i, 'Close'] = y.loc[i, 'Close'] - y.loc[i - 1, 'Close']
    
# 3. Set the first change of Close price as mean value
y1.loc[0, 'Close'] = 0.055434
y1.mean()

# 4. Plot the relationship between x and y1
plt.figure(figsize=(8, 8))

plt.scatter(x, y1['Close'], alpha=0.2)
plt.xlabel('Maximum increase of the price in a day')
plt.ylabel('Increae of price comapred to yesterday');

# We can know the facts that when the x value is large, then it's more likely to have a larger y value.
# When the x is larger than 5, then it's more likely to have positive y value which means the higher Close price compared to yesterday.
# But this is relatively obvious result because when the maximum rise is high, then the Close price will also high.
# In order to make a profit, we need to predict the Ctock price of the future. 
# So, we will analyze the relationship between today's maximum rise and tomorrow's Close price not today's Close price.
```

```python
# 5. Analyze more, Today's High rise → Impact on tomorrow's price

y2 = y1.loc[1:, 'Close']
y2 = y2.reset_index().drop(columns = ['index'])
y2.mean() # >>> 0.055444

y2 = y2.append({'Close':0.055444}, ignore_index=True)
# x, y2
plt.figure(figsize=(8, 8))

plt.scatter(x, y2['Close'], alpha=0.2)
plt.xlabel('Maximum increase of the price in a day')
plt.ylabel("Increase of tomorrow's close price compared to today")
plt.show();

# It seems that there is no evident relationship between x and y.
# Let's take a closer look at the data.
```

```python
# Let's focus on the data of y2 at x = 5.0.

indices = x[(4.8 < x['Rise']) & (x['Rise'] < 5.2)].index.tolist()
x1 = x.loc[indices, 'Rise'].tolist()
y3 = y2.loc[indices, 'Close'].tolist()

plt.hist(y3, bins=25);
# As you can check in the histogram and the previous scatter plot, it doesn't have a meaningful pattern.
# Thus, we can conclude that it's difficult to predict tomorrow's Close price from the today's maximum rise.
```

```python
# We can also check the result in the probability density function.
# It is almost symmetric.

from scipy import stats
import math

mu = y2.loc[indices, 'Close'].mean()
variance = y2.loc[indices, 'Close'].var()
sigma = math.sqrt(variance)
xx = np.linspace(-25, 25, 200)
plt.plot(xx,stats.norm.pdf(xx, mu, sigma))
```
